```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
DATA_DIR = "../data/ml-100k"
ratings_file = os.path.join(DATA_DIR, "u.data")

# Column names per MovieLens documentation
columns = ["user_id", "item_id", "rating", "timestamp"]
ratings = pd.read_csv(ratings_file, sep="\t", names=columns)

print(ratings.head())
print(ratings.describe())

# Let's see how many unique users and items
print("Number of unique users: ", ratings["user_id"].nunique())
print("Number of unique items: ", ratings["item_id"].nunique())

# Distribution of ratings
ratings["rating"].value_counts().sort_index().plot(kind="bar", title="Rating Distribution")
```
